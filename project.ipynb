```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
import seaborn as sns 
```

```python
url = 'https://raw.githubusercontent.com/edensayag/project/main/life_expectancy.csv' 
ex_df = pd.read_csv(url)
ex_df
```

```python
developing_country= ex_df[ex_df.loc[:,'Status'] == 'Developing']
developing_country 
#מדינות שמתפתחות
```

```python
developed_country=ex_df[ex_df.loc[:,'Status'] == 'Developed']
developed_country
#מדינות מפותחות
```

```python
sum_hpb1 = developing_country.groupby(['Year','Status'])[['Hepatitis B']].sum()
sum_hpb1
#סכום הפטיטס ב במדיונות שמתפתחות
```

```python
sum_hpb1['Hepatitis B'].mean()
#סהכ
```

```python
sum_hpb2= developed_country.groupby(['Year','Status'])[['Hepatitis B']].sum()
sum_hpb2
#סכום הפטיטס ב במדיונות שמפותחות
```

```python
sum_hpb2['Hepatitis B'].sum()
#סהכ
```

```python
fig, ax = plt.subplots(1,2, figsize = (5,4))
plt.subplots_adjust(wspace = 1)
sns.boxplot(data=sum_hpb1, y= 'Hepatitis B', ax=ax[0])
sns.boxplot(data=sum_hpb2, y= 'Hepatitis B', ax=ax[1])


plt.show()#
```
