```python
fig, ax = plt.subplots(1,2, figsize = (14,8))
plt.subplots_adjust(wspace = 0.3)
sns.histplot(x='Year',y='Life.expectancy', data=ex_df2, stat='probability', bins=16,hue="rank", ax = ax[0])
sns.histplot(x='Year',y='Life.expectancy', data=ex_df2, stat='probability', bins=16 ,hue="stage", ax = ax[1])
sns.move_legend(ax[0], "upper center", bbox_to_anchor=(0.1,1.1))
sns.move_legend(ax[1], "upper right", bbox_to_anchor=(0.1,1.1))
plt.show()
```

```python
plt.figure(figsize=(5,3))
ax = sns.lineplot(x="Year", y="Income.composition.of.resources", legend=False,color="turquoise", data=ex_df2,errorbar=None)
sns.lineplot(x="Year", y="Life.expectancy", ax=ax.twinx(), legend=False, color="magenta", data=ex_df2,errorbar=None)
ax.figure.legend(handles=[Line2D([], [], marker='', color="magenta", label='Life expectancy'), Line2D([], [], marker='', color="turquoise", label='income composition of resources')], bbox_to_anchor=(1.04, 0.5), loc="center left")
plt.show()
```

```python
plt.figure(figsize=(5,3))
ax = sns.lineplot(x="Year", y="thinness..1.19.years", legend=False,color="turquoise", data=ex_df2[ex_df2['rank']=='high pop'],errorbar=None)
sns.lineplot(x="Year", y="Life.expectancy", ax=ax.twinx(), legend=False, color="magenta", data=ex_df2[ex_df2['rank']=='high pop'],errorbar=None)
ax.figure.legend(handles=[Line2D([], [], marker='', color="magenta", label='Life expectancy'), Line2D([], [], marker='', color="turquoise", label='thinness')], bbox_to_anchor=(1.04, 0.5), loc="center left")
plt.show()
```
