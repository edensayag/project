### Here, we created a new column divided into three groups: low income , medium income, and high income.

```python
ex_df2['stage']=pd.cut(ex_df2["Income.composition.of.resources"],bins=3,labels=("low income","medium income","high income" ))
x=ex_df2.groupby('stage')["Income.composition.of.resources"].describe()
x
```

#### We compared the groups we created and conducted scatterplots to observe the probability.

```python
plt.figure(figsize=(8,6))
ax =sns.scatterplot(data=ex_df2,x="Year",y="Life.expectancy", hue="stage",palette = 'colorblind')
style = dict(size=12, color='red')
sns.move_legend(ax, "upper center", bbox_to_anchor=(1.3,0.6))
ax.grid(False)
plt.xticks(range(2000, 2016))
plt.show()
```

```python
plt.figure(figsize=(8,6))
ax = sns.lineplot(x="Year", y="Income.composition.of.resources", legend=False, color="turquoise", data=ex_df2, errorbar=None)
sns.lineplot(x="Year", y="Life.expectancy", ax=ax.twinx(), legend=False, color="magenta", data=ex_df2, errorbar=None)
ax.figure.legend(handles=[Line2D([], [], marker='', color="magenta", label='Life expectancy'), Line2D([], [], marker='', color="turquoise", label='income composition of resources')], bbox_to_anchor=(1.04, 0.5), loc="center left")
plt.xticks(range(2000, 2016))
ax.grid(False) # Remove grid
plt.show()
```

#### Here, we observed the country of Grenada, which boasts a higher life expectancy compared to other low-income countries, as well as greater educational opportunities. This prompted us to wonder what makes Grenada so special.

```python

# Define the color palette for low income countries from the first graph
color_palette_low = sns.color_palette("Set2")[:len(order_count)]

# Create a dictionary to map each country to its corresponding color in the first graph
color_map_low = dict(zip(order_count, color_palette_low))

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(12, 6))
plt.style.use('_mpl-gallery')

# Code for the first graph
x1 = range(1, 6)
y1 = clow.groupby(['Country'])['Life.expectancy'].mean()
sorted_data1 = sorted(zip(x1, y1.items()), key=lambda item: item[0])
x_sorted1, y_sorted1 = zip(*sorted_data1)
for i, (x_val, (country, y_val)) in enumerate(zip(x_sorted1, y_sorted1)):
    ax1.plot([x_val, x_val], [0, y_val], linestyle='-', color=color_map_low[country])
    ax1.scatter(x_val, y_val, marker='.', s=80, color=color_map_low[country])
ax1.set(xlim=(0, 6), xticks=np.arange(1, 6),
        ylim=(0, 80), yticks=np.arange(0, 80, 10))  # Changing the yticks to intervals of 10
ax1.set_xticklabels(order_count, rotation=80)
ax1.set_title('life expectancy per country')

# Code for the second graph
x2 = range(1, 6)
y2 = clow.groupby(['Country'])['Schooling'].mean()
sorted_data2 = sorted(zip(x2, y2.items()), key=lambda item: item[0])
x_sorted2, y_sorted2 = zip(*sorted_data2)
for i, (x_val, (country, y_val)) in enumerate(zip(x_sorted2, y_sorted2)):
    ax2.plot([x_val, x_val], [0, y_val], linestyle='-', color=color_map_low[country])
    ax2.scatter(x_val, y_val, marker='.', s=80, color=color_map_low[country])
ax2.set(xlim=(0, 6), xticks=np.arange(1, 6),
        ylim=(0, 17), yticks=np.arange(0, 17))  # Changing the yticks to intervals of 10
ax2.set_xticklabels(order_count, rotation=80)
ax2.set_title('schooling per country')

# Code for the third graph
x3 = range(1, 6)
y3 = clow.groupby(['Country'])['thinness..1.19.years'].mean()
sorted_data3 = sorted(zip(x3, y3.items()), key=lambda item: item[0])
x_sorted3, y_sorted3 = zip(*sorted_data3)
for i, (x_val, (country, y_val)) in enumerate(zip(x_sorted3, y_sorted3)):
    ax3.plot([x_val, x_val], [0, y_val], linestyle='-', color=color_map_low[country])
    ax3.scatter(x_val, y_val, marker='.', s=80, color=color_map_low[country])
ax3.set(xlim=(0, 6), xticks=np.arange(1, 6),
        ylim=(0, 18), yticks=np.arange(0, 18))  # Changing the yticks to intervals of 10
ax3.set_xticklabels(order_count, rotation=80)
ax3.set_title('thinness percent per country')

plt.tight_layout()
plt.show()
```

#### Here, we researched only the low-income country.

#### From the graph, "Grenada" has one of the lowest percentages of thinness and the highest number of years of schooling. "Turkmenistan" has the lowest thinness rate and is only 6 years apart in terms of schooling from "Grenada," which leads us to indicate that the more you learn in the educational system, the greater your life expectancy.
